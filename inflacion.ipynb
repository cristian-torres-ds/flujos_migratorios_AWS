```python
import pandas as pd
import numpy as np
```

CSV

```python
df_inf = pd.read_csv('Data\API_FP.CPI.TOTL.ZG_DS2_es_csv_v2_5689785.csv', skiprows = 3)
```

```python
df_inf.head(3)
```

```python
df_inf = df_inf[df_inf["Indicator Name"] == "Inflación, precios al consumidor (% anual)"]
```

```python
df_inf.drop(columns = 'Indicator Code', inplace = True)
```

```python
paises = ['India','México','China','Filipinas','Pakistán']

inf_filter = df_inf[df_inf['Country Name'].isin(paises)]
```

```python
inf_filter = inf_filter.reset_index(drop=True)
```

```python
cols_to_drop = [str(year) for year in range(1960, 2000)]
inf_filter = inf_filter.drop(cols_to_drop, axis=1)
```

```python
inf_filter.dropna(axis=1, how='all',inplace = True)
```
