```python
import pandas as pd
import numpy as np
```

CSV

```python
df_pib = pd.read_csv('Data\API_NY.GDP.MKTP.CD_DS2_es_csv_v2_5721684.csv', skiprows=3)
```

```python
df_pib.head(4)
```

```python
df_pib = df_pib[df_pib["Indicator Name"] == "PIB (US$ a precios actuales)"]
```

```python
df_pib.drop(columns = 'Indicator Code', inplace = True)
```

```python
paises = ['India','México','China','Filipinas','Pakistán']

pib_filter = df_pib[df_pib['Country Name'].isin(paises)]
```

```python
pib_filter = pib_filter.reset_index(drop=True)
```

```python
cols_to_drop = [str(year) for year in range(1960, 2000)]
pib_filter = pib_filter.drop(cols_to_drop, axis=1)
```

```python
pib_filter = pib_filter.applymap(lambda x: '{:.2f}'.format(x) if isinstance(x, float) else x)
```

```python
pib_filter.dropna(axis=1, how='all',inplace = True)
```

```python
pib_filter.drop('Unnamed: 67', axis=1, inplace=True)
```
