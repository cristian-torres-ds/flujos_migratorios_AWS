```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data\MIG_31082023183055443.csv')
```

```python
df.head(10)
```

```python
df = df[df["Variable"] == "Inflows of foreign population by nationality"]
```

```python
df = df.drop(columns= ['VAR','GEN','Gender','YEA','Flag Codes','Flags'])
```

```python
df_pivot = df.pivot_table(index=['Country of birth/nationality','Country'], columns='Year', values='Value')
```

```python
df_pivot.head(10)
```

```python
countries = ['India','Mexico','China','Philippines','Pakistan']
```

```python
df_filtered = df_pivot.loc[countries]
```

```python
df_filtered.head(10)
```

```python
df_filtered.dropna(inplace = True)
```

HACER UN TOP 5 DEL ULTIMO AÑO

```python
df3 = df_filtered.copy()
```

```python
total_migrantes = df3.sum(axis=1)
```

```python
total_migrantes
```

```python
# Obtiene el top 5 de países de destino para cada país de origen
top_5 = total_migrantes.groupby('Country of birth/nationality').apply(lambda x: x.nlargest(5).index)
```

```python
mask = pd.Series(False, index=df3.index)
for pais_origen, paises_destino in top_5.items():
    mask |= ((df3.index.get_level_values('Country of birth/nationality') == pais_origen) & 
             df3.index.get_level_values('Country').isin(paises_destino.get_level_values(1)))
df3_filtrado = df3[mask]
```
